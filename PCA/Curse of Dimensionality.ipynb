```python
import random
import math
import numpy as np
from scipy.spatial import distance
import pandas as pd
```

```python
n1=[]
d1=[]
gamma=[]
for i in range(30):
    n=random.randint(100,1000)
    d=random.randint(1,100)
    matrix=np.random.randint(100,size=(n,d))
    i=0
    ematrix=[]
    while(i<len(matrix)):
        for j in range(i+1,len(matrix)):
            mat=matrix[i]
            dst = distance.euclidean(mat, matrix[j])
            ematrix.append(dst)
        i=i+1
    minval=min(ematrix)
    maxval=max(ematrix)
    gamma1=math.log((maxval-minval)/minval)
    n1.append(n)
    d1.append(d)
    gamma.append(gamma1)
```

```python
percentile_list = pd.DataFrame(
    {'lst1Title': n1,
     'lst2Title': d1,
     'lst3Title': gamma
    })
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
df = pd.DataFrame({'x': percentile_list.lst1Title, 'y': percentile_list.lst2Title, 'z': percentile_list.lst3Title}, index=range(len(percentile_list)))

# Plot using `.trisurf()`:
ax.set_xlabel('n')
ax.set_ylabel('d')
ax.set_zlabel('gamma');
ax.plot_trisurf(df.x, df.y, df.z, cmap=cm.jet, linewidth=0.2)
plt.show()
```

```python
from numpy import linalg as LA
a1=[]
b1=[]
gamma3=[]
for i in range(30):
    a=random.randint(100,1000)
    b=random.randint(1,100)
    matrix=np.random.randint(100,size=(a,b))
    dst = LA.norm(matrix,axis=1)
    minval=min(dst)
    maxval=max(dst)
    gamma2=math.log((maxval-minval)/minval)
    a1.append(a)
    b1.append(b)
    gamma3.append(gamma2)
```

```python
percentile_list1 = pd.DataFrame(
    {'lst1Title': a1,
     'lst2Title': b1,
     'lst3Title': gamma3
    })
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = Axes3D(fig)
df = pd.DataFrame({'x': percentile_list1.lst1Title, 'y': percentile_list1.lst2Title, 'z': percentile_list1.lst3Title}, index=range(len(percentile_list)))

# Plot using `.trisurf()`:
ax.set_xlabel('n')
ax.set_ylabel('d')
ax.set_zlabel('gamma');
ax.plot_trisurf(df.x, df.y, df.z, cmap=cm.jet, linewidth=0.2)
plt.show()
```
